```python
def compare_imputation(df_test, df_after, column_name):
    missing_indices={column_name:df_test[df_test[column_name].isnull()].index.to_list() for column_name in df_test.columns if df_test[column_name].isnull().any()}

    if column_name not in df_test.columns:
        print(f"No missing values were found in the column '{column_name}'")
        return
    
    idx_list=missing_indices[column_name]

    comparision=pd.DataFrame({
        "Before" : df_test.loc[idx_list, column_name],
        "After" : df_after.loc[idx_list, column_name]
        })
    print(f"Changes in column :{column_name}")
    return comparision


compare_imputation(df_test, df_after, "A")
```

```python
def detect_outliers_iqr(data, col):
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-(1.5*IQR)
    upper=Q3+(1.5*IQR)
    outliers=data[(data[col]<lower) | (data[col]>upper)]
    return outliers[[col]]
```

```python
def exclude_outliers(data, col):
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    upper=Q3+(1.5*IQR)
    lower=Q1-(1.5*IQR)
    req_data=data[(data[col]>lower) & (data[col]<upper)]
    return req_data[[col]]
```

```python
def forward_selection(X, Y, threshold=0.05):
    model_summary_captured=False
    selected_features=[]
    remaining_features=list(X.columns)
    while remaining_features:

        pvals={}
        for feature in remaining_features:
            model=sm.OLS(Y, sm.add_constant(X[selected_features+[feature]])).fit()
            if not model_summary_captured:
                model_summary=sm.OLS(Y, X).fit()
                print("Model summary has been captured")
                print("**"*30)
                print(model_summary.summary())
                model_summary_captured=True
            p_values=model.pvalues[feature]
            pvals[feature]=p_values
        pvalue_min_feature=min(pvals, key=pvals.get)
        min_pval=pvals[pvalue_min_feature]
        if min_pval<threshold:
            selected_features.append(pvalue_min_feature)
            remaining_features.remove(pvalue_min_feature)
            print(f"Added {pvalue_min_feature} with p-value {pvals[pvalue_min_feature].round(4)}")
        else:
            break
    return selected_features

```
